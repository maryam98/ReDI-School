## The dataset:
The datasets contains information about rental appartments in Berlin. 
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat has?

```python
import pandas as pd
import numpy as np
import datetime

# this is a hack to allow displaying more than one result per notebook cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# reading data
rentals = pd.read_csv("https://raw.githubusercontent.com/juliandnl/redi_ss20/master/berlin_rental.csv")
rentals.head()
```

### 1. Add 200 Euros to the rent price of all apartments in Mitte.



```python
rentals.loc[rentals.Region=='Mitte']
```

```python
rentals.loc[rentals.Region=='Mitte',['Rent']]=rentals.Rent+200
```

```python
rentals.loc[rentals.Region=='Mitte']
```

### 2. Add a variable to the dataset that shows the rent price in U.S. Dollars and another variable that shows the space in square feet, instead of square meters.

```python
rentals['Dollr']=round(rentals.Rent*1.12)
rentals['feet']=rentals.Space*0.092903
```

```python
rentals
```

### 3. Create a flag to indicate that the apartment is being rented for the first time (tip: look in 'Conditions')

```python
rentals['FirstTimeRent']='No'
rentals.loc[rentals.Condition=='first_time_use','FirstTimeRent']='Yes'
```

```python
rentals
```

### 4. Remove from the dataset all apartments from 'Kreuzberg'

```python
rentals.drop(rentals.index[rentals.Region=='Köpenick'],inplace=True)
```

```python
rentals.Region.unique()
```

### 5. Create a new dataset that shows the average price per Region. Then, join this information back into the main dataset (tip: use pd.merge())

```python
new_data=rentals.groupby('Region')['Rent'].mean().reset_index()
```

```python
new_data
```

### 6. Calculate, for each apartment, the difference between its price and the average price of the Region. Create a flag to indicate if this apartment is more expensive than the average.

```python
rentals=rentals.merge(new_data,how='left',on='Region',suffixes=('','_mean'))
```

```python
rentals
```

```python
diff_Rent=rentals.Rent_mean-rentals.Rent
```

```python
rentals['Expensive']='No'
rentals.loc[rentals.Rent > rentals.Rent_mean,['Expensive']]='Yes'
```
