## The dataset:
The datasets contains information about rental appartments in Berlin. 
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat has?

```
# Imports
import pandas as pd
```

```
# Read in dataset
rentals = pd.read_csv("https://raw.githubusercontent.com/juliandnl/redi_ss20/master/berlin_rental.csv")
rentals.head()
```

# 1.How many apartments is in **Mitte** in a **'well-kept'** condition?

```
#1_1:
aaa=rentals[(rentals.Region=="Mitte") & (rentals.Condition=="well_kept")]['Region'].count()
rentals['Number']=aaa
rentals[(rentals.Region=="Mitte") & (rentals.Condition=="well_kept")].groupby(['Region','Condition'])[['Number']].count()
```

# 2.How many apartments are there in **Kreuzberg** built **before 1900**?

```
# 2_1:
rentals[(rentals.Region=="Kreuzberg")&(rentals.Year_Construction<1900)]
```

```
#2_2:
rentals[(rentals.Region=="Kreuzberg")&(rentals.Year_Construction<1900)]['Region'].count()
```

```
#2_3:
rentals[(rentals.Region=="Kreuzberg")&(rentals.Year_Construction<1900)].shape[0]
```

# 3.What are the average **space**s of the apartments by the number of **rooms**?

```
# your code

rentals.groupby('Rooms')[['Space']].mean()
```

# 4.Which is the most expensive **region** on average and what is the maximum **rent** for this region?

```
print('****average rent for each region:****')
print(rentals.groupby('Region')[['Rent']].mean())
print('\n****the most expensive region on average****')
a=rentals.groupby('Region')[['Rent']].mean()
ab=a.loc[a.Rent==a.Rent.max(),['Rent']]
print(ab)
print('\n**** what is the maximum rent for region',ab.index[0],'  ****')
rentals[rentals.Region==ab.index[0]]['Rent'].max()
```

# Bonus Question:
5. What are the min, max and average number of rooms of the apartments in Köpenick?


```
#5_1:
rentals[rentals.Region=='Köpenick'][['Rooms']].agg(['min','max','mean'])
```

```
#5_2:
rentals.loc[rentals.Region=='Köpenick','Rooms'].agg(['min','max','mean']).to_frame()
```

```
#5_3:
rentals[(rentals.Region=='Köpenick')][['Region','Rooms']].groupby('Region')[['Rooms']].agg(['min','max','mean'])
```

```
#5_4:
rentals.loc[rentals.Region=='Köpenick',['Region','Rooms']].groupby('Region')[['Rooms']].agg(['min','max','mean'])
```

```
#5_5:
rentals.groupby('Region' )[['Rooms']].agg(['min','max', 'mean']).loc[['Köpenick']]
```
